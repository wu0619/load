```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('load.csv')
load_col = 'out.site_energy.total.energy_consumption.kwh'
```

```python
data['timestamp'] = pd.to_datetime(data['timestamp'])
data[load_col] = data[load_col] * 4 / 1e3
```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.figure(figsize=(12, 6))
plt.title('Load Curve of Field A in 2023', fontsize=18)
plt.plot(data[load_col])
plt.xlabel('Data Count', fontsize=14)
plt.ylabel('Load (kW)', fontsize=14)
plt.tight_layout()
# plt.savefig('field.png', dpi=1300)
plt.show()
```

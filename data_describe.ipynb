```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
field_df = pd.read_csv('load.csv')
steel_df = pd.read_csv('Steel_industry_data.csv')
```

```python
field_df['Load'] = field_df['Load'] * 4
field_df.describe()
```

```python
steel_df['Usage_kWh'] = steel_df['Usage_kWh'] * 4
steel_df.describe()
```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.figure(figsize=(12, 6))
plt.title('Load Curve of Field A in 2023', fontsize=18)
plt.plot(field_df['Load'])
plt.xlabel('Data Count', fontsize=14)
plt.ylabel('Load (kW)', fontsize=14)
plt.tight_layout()
plt.savefig('field.png', dpi=1300)
plt.show()
```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.figure(figsize=(12, 6))
plt.title('Load Curve of DAEWOOL Steel Industry in 2018', fontsize=18)
plt.plot(steel_df['Usage_kWh'])
plt.xlabel('Data Count', fontsize=14)
plt.ylabel('Load (kW)', fontsize=14)
plt.tight_layout()
plt.savefig('steel.png', dpi=1300)
plt.show()
```
